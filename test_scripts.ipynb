
This script run the EcoMotionZip on videos and records the data

```python
import pandas as pd

# FILEPATH: /Users/mrat0010/Documents/GitHub/EcoMotionZip/test_scripts.ipynb
import os
# os.environ["IMAGEIO_FFMPEG_EXE"] = "/usr/bin/ffmpeg"
from moviepy.editor import VideoFileClip

def get_video_info(directory):
    video_info = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".mp4") or file.endswith(".avi") or file.endswith(".mkv"):
                file_path = os.path.join(root, file)
                video = VideoFileClip(file_path)
                duration = video.duration
                fps = video.fps
                resolution = video.size
                file_size = os.path.getsize(file_path)
                video_info.append({
                    "file_name": file, 
                    "file_size": file_size,
                    "duration": duration,
                    "fps": fps,
                    "resolution": resolution
                })
    return pd.DataFrame(video_info)

directory = "./testing/test_videos"
df = get_video_info(directory)
df['file_size'] = df['file_size'] / 1048576
df
```

```python
import os
import pandas as pd

directory = "./testing/EcoMotionZip"

video_info = []
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith(".mp4") or file.endswith(".avi") or file.endswith(".mkv"):
            file_path = os.path.join(root, file)
            video = VideoFileClip(file_path)
            duration = video.duration
            fps = video.fps
            file_size = os.path.getsize(file_path) / 1048576  # Convert to MB
            log_file_path = os.path.join(root, f"{os.path.splitext(file)[0]}_output.log")            
            if os.path.exists(log_file_path):
                with open(log_file_path, "r") as log_file:
                    lines = log_file.readlines()
                    last_line = lines[-1].strip()
                    if last_line.endswith("seconds."):
                        processing_time = last_line.split()[-2]
                    else:
                        processing_time = None
            else:
                processing_time = None
            video_info.append({
                "Video Name": file,
                "File Size (MB)": file_size,
                "Duration": duration,
                "FPS": fps,
                "Processing Time": processing_time
            })

df_output = pd.DataFrame(video_info)
df_output
```

```python
merged_df = df.merge(df_output, left_on='file_name', right_on='Video Name', how='inner')

merged_df['Percentage Reduction'] = (merged_df['file_size'] - merged_df['File Size (MB)']) / merged_df['file_size'] * 100

merged_df
```
