```python
import pandas as pd

# FILEPATH: /Users/mrat0010/Documents/GitHub/EcoMotionZip/test_scripts.ipynb
import os
# os.environ["IMAGEIO_FFMPEG_EXE"] = "/usr/bin/ffmpeg"
from moviepy.editor import VideoFileClip

def get_video_info(directory):
    video_info = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".mp4") or file.endswith(".avi") or file.endswith(".mkv"):
                file_path = os.path.join(root, file)
                video = VideoFileClip(file_path)
                duration = video.duration
                fps = video.fps
                resolution = video.size
                file_size = os.path.getsize(file_path)
                frame_count = video.reader.nframes
                video_info.append({
                    "Video Name": file, 
                    "file_size": file_size,
                    "duration": duration,
                    "frame_count": frame_count,
                    "fps": fps,
                    "resolution": resolution
                    
                })
    return pd.DataFrame(video_info)

directory = "./testing/test_videos"
df = get_video_info(directory)
df['file_size'] = df['file_size'] / 1048576
df
```

```python
import os
import pandas as pd

directory = "./testing/EcoMotionZip_RPi5"

video_info = []
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith(".mp4") or file.endswith(".avi") or file.endswith(".mkv"):
            file_path = os.path.join(root, file)
            video = VideoFileClip(file_path)
            duration = video.duration
            fps = video.fps
            file_size = os.path.getsize(file_path) / 1048576  # Convert to MB
            frame_count = video.reader.nframes
            log_file_path = os.path.join(root, f"{os.path.splitext(file)[0]}_output.log")            
            if os.path.exists(log_file_path):
                with open(log_file_path, "r") as log_file:
                    lines = log_file.readlines()
                    last_line = lines[-1].strip()
                    if last_line.endswith("seconds."):
                        processing_time = last_line.split()[-2]
                    else:
                        processing_time = None
            else:
                processing_time = None
            video_info.append({
                "Video_Name_RPi5": file,
                "File_Size_(MB)_RPi5": file_size,
                "Duration_RPi5": duration,
                "Frame_Count_RPi5": frame_count,
                "FPS_RPi5": fps,
                "Processing_Time_RPi5": processing_time
            })

df_RPi5 = pd.DataFrame(video_info)

df_RPi5
```

```python
# Merge contents of df and df_Mac using the Video_Name column as the key. Ignore the extension in the filename when merging.
# Merge the dataframes
df_merged = pd.merge(df, df_RPi5, left_on=df['Video Name'].str.split('.').str[0], right_on=df_RPi5['Video_Name_RPi5'].str.split('.').str[0], how='inner')

# Remove the duplicate key column, Video_Name_Mac and FPS_Mac columns
df_merged = df_merged.drop(columns=['key_0', 'Video_Name_RPi5'])

# Rename the Duration_RPi4 and FPS_RPi4 columns to Duration_Processed and FPS_Processed
df_merged = df_merged.rename(columns={'Duration_RPi4': 'Duration_Processed', 'FPS_RPi4': 'FPS_Processed'}) 

# Calclulate the percentage reduction in file size by comparing file_size and File_Size_(MB)_Mac columns
df_merged['File_Size_Reduction_(%)'] = (df_merged['file_size'] - df_merged['File_Size_(MB)_RPi5']) / df_merged['file_size'] * 100

# Calclulate the percentage reduction in frame count by comparing frame_count and Frame_Count columns
df_merged['Frame_Count_Reduction_(%)'] = (df_merged['frame_count'] - df_merged['Frame_Count_RPi5']) / df_merged['frame_count'] * 100

# Add the results from the RPi5 to the merged dataframe
# df_merged = pd.merge(df_merged, df_RPi5, left_on=df_merged['Video Name'].str.split('.').str[0], right_on=df_RPi5['Video_Name_RPi5'].str.split('.').str[0], how='inner')

# Remove the columns on RPi keeping only processing time
# df_merged = df_merged.drop(columns=['key_0', 'Video_Name_RPi5', 'Duration_RPi5', 'FPS_RPi5'])

# Add the results from the RPi4 to the merged dataframe simimar to the RPi5
df_merged = pd.merge(df_merged, df_RPi4, left_on=df_merged['Video Name'].str.split('.').str[0], right_on=df_RPi4['Video_Name_RPi4'].str.split('.').str[0], how='inner')

# Remove the columns on RPi keeping only processing time
df_merged = df_merged.drop(columns=['key_0', 'Video_Name_RPi4', 'Duration_RPi4', 'FPS_RPi4'])

df_merged = df_merged.sort_values(by=['Video Name'])

# Round of values to 2 decimal places
df_merged = df_merged.round(decimals=2)

# Read the transfer times data from the CSV file
transfer_times = pd.read_csv('./testing/transfer_times.csv')

# Merge the transfer times data with the existing dataframe
df_merged = pd.merge(df_merged, transfer_times, left_on='Video Name', right_on='File', how='left')

# Remove the duplicate key column, File column and rename the Transfer Time column as TT_Original
df_merged = df_merged.drop(columns=['File'])
df_merged = df_merged.rename(columns={'Transfer Time': 'TT_Original'})

# round off TT_Original to 2 decimal places
df_merged['TT_Original'] = df_merged['TT_Original'].round(decimals=2)


df_merged

```

```python
# Creaate a dataframe with the results of the ./testing/transcode_times_RPi5.csv file
TT = pd.read_csv('./testing/transfer_times_RPi5.csv')

# breake the contents of the File column to seperate columns seperated by /
TT[['Folder Name', 'File Name']] = TT.File.str.split("/", expand=True)

# Calculate the sum of Transfer Times for each Folder Name and insert them to a new dataframe
TT_RPi5 = TT.groupby(['Folder Name']).sum()

#remove columns File and File Name
TT_RPi5 = TT_RPi5.drop(columns=['File', 'File Name'])

# Merge the TT_RPi5 data with the df_merged dataframe. Use the Folder Name column as the key and ignore the extension in the filename when merging.
df_merged = pd.merge(df_merged, TT_RPi5, left_on=df_merged['Video Name'].str.split('.').str[0], right_on=TT_RPi5.index.str.split('.').str[0], how='inner')

# Remove the duplicate key column, File column and rename the Transfer Time column as TT_RPi5
df_merged = df_merged.drop(columns=['key_0'])
df_merged = df_merged.rename(columns={'Transfer Time': 'TT_RPi5'})

# round off TT_RPi5 to 2 decimal places
df_merged['TT_RPi5'] = df_merged['TT_RPi5'].round(decimals=2)

# # Calculate the percentage reduction in transfer time by comparing TT_Original and TT_RPi5 columns
df_merged['TT_Reduction_(%)'] = (df_merged['TT_Original'] - df_merged['TT_RPi5']) / df_merged['TT_Original'] * 100

# round off TT_Reduction_(%) to 2 decimal places
df_merged['TT_Reduction_(%)'] = df_merged['TT_Reduction_(%)'].round(decimals=2)



df_merged
```
